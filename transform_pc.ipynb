```python
    def pc_crs(pc1, pc2):
        with laspy.open(pc1) as las:
            hdr = las.header
            crs_from = hdr.parse_crs()
        log.debug(f"CRS used is {crs_from}")
        with laspy.open(pc2) as las:
            hdr = las.header
            crs_to = hdr.parse_crs()
        log.debug(f"CRS used is {crs_to}")
        if not crs_from or not crs_to:
            raise ValueError("One or both LAS files are missing CRS information.")
        
        transformer = Transformer.from_crs(crs_from, crs_to, always_xy=False)
        log.debug(f"Transformer created between {crs_from} and {crs_to}")
        return transformer
    
    pc_crs(pc1, pc2)
    
```
