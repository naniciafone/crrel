```python
# good docs on types of filters used: https://pdal.io/stages/filters.html#ground-unclassified

input_laz = "C:/Users/RDCRLSMC/Desktop/snotelrodeo_04042025_redo/09_EXPORT/canyon_peak/SnotelRodeo20250404250404_154743_Scanner_1.laz"

print(input_laz)
```

```python
pipeline_json = {
    "pipeline": [
        {"type": "readers.las", "filename": input_laz}
    ]
}

pipeline = pdal.Pipeline(json.dumps(pipeline_json))
count = pipeline.execute()
print("Point count:", count)
```

```python
pipeline_json = {
    "pipeline": [
        {
            "type": "readers.las",
            "filename": input_laz
        },
        {
            "type": "filters.mongo",
            "expression": {
                "$and": [
                    {"ReturnNumber": {"$gt": 0}},
                    {"NumberOfReturns": {"$gt": 0}}
                ]
            }
        }
    ]
}

# Convert dict to JSON string
pipeline_text = json.dumps(pipeline_json, indent=4)

# Run pipeline
pipeline = pdal.Pipeline(pipeline_text)
count = pipeline.execute()

print("Input file:", input_laz)
print("Points after filtering:", count)
```
