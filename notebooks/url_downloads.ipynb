```python
from urllib.request import urlretrieve
import os
```

```python
download_dir = "C:/Users/RDCRLSMC/Desktop/canyonpeak/REF_LAZ/"
url_list = os.path.join(download_dir, "downloadlist.txt")
os.makedirs(download_dir, exist_ok=True) # Creates the directory if it doesn't exist
```

```python
# Open and read the file
with open(url_list, 'r') as file:
    urls = [line.strip() for line in file if line.strip()]  # Read and strip each line

# Output the list of URLs
print("Loaded URLs:")
for url in urls:
    print(url)
```

```python
for url in urls:
    filename = os.path.basename(url)  # Extract filename from URL
    local_filepath = os.path.join(download_dir, filename)
    try:
        urlretrieve(url, local_filepath)
        print(f"Downloaded: {filename} to {download_dir}")
    except Exception as e:
        print(f"Error downloading {url}: {e}")
```
