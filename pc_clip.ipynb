```python
def clip_point_cloud(input_file, output_file, polygon_file):
    pipeline = [
        {
            "type": "readers.las", 
            "filename": input_file
        },
        {
            "type": "filters.crop", 
            "polygon": polygon_file
        },
        {
            "type": "writers.las", 
            "filename": output_file
        }
    ]
    pdal_pipeline = pdal.Pipeline(json.dumps(pipeline))
    pdal_pipeline.execute()
```
